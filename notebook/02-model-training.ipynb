```python
import subprocess

def cmd(command_string):
    """Executes a raw string directly in the system shell."""
    result = subprocess.run(command_string, shell=True, capture_output=True, text=True)
    
    if result.returncode == 0:
        print(result.stdout)
    else:
        print("Error:\n", result.stderr)

# Usage:
# cmd("echo 'Hello World' > test.txt && cat test.txt")
```
